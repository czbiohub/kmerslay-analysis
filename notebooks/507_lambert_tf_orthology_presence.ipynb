```python
import pandas as pd
import seaborn as sns
```

# Read TF presence/absence across species from Lambert et al

http://humantfs.ccbr.utoronto.ca/download.php

```python
tf_orthologs_lambert = pd.read_csv("http://humantfs.ccbr.utoronto.ca/download/v_1.01/Fig3A.csv", index_col=0)
tf_orthologs_lambert = tf_orthologs_lambert.fillna(0)
print(tf_orthologs_lambert.shape)
tf_orthologs_lambert.head()
```

```python
tf_orthologs_lambert.loc[tf_orthologs_lambert['Mouse Lemur'] == 0]
```

```python
tf_orthologs_lambert.loc[(tf_orthologs_lambert['Mouse'] == 0) & (tf_orthologs_lambert['Mouse Lemur'] > 0)]
```

```python
sns.jointplot('Mouse', "Mouse Lemur", tf_orthologs_lambert)
```

```python
tf_names = pd.DataFrame(tf_orthologs_lambert.index.str.split('#', expand=True).tolist(), columns=['symbol', 'ensembl_id'])
print(tf_names.shape)
tf_names.head()
```
